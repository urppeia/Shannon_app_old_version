```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/shares/grossniklaus.botinst.uzh/eharputluoglu/accessions.csv')

# Group the data by Country and Sequenced by, then count occurrences
country_sequence_counts = data.groupby(['Country', 'Sequenced by']).size().reset_index(name='Counts')

# Pivot the data to wide format
pivot_data = country_sequence_counts.pivot(index='Country', columns='Sequenced by', values='Counts').fillna(0)

# Convert back to long format
long_data = pivot_data.reset_index().melt(id_vars='Country', var_name='Sequenced_by', value_name='Counts')

# Set up the matplotlib figure with increased size
plt.figure(figsize=(14, 10))

# Create a bar plot using seaborn with increased font sizes
sns.barplot(data=long_data, x='Country', y='Counts', hue='Sequenced_by', palette='viridis')

# Rotate x-axis labels for better readability and increase font size
plt.xticks(rotation=45, ha='right', fontsize=12)

# Increase the font size for y-axis labels
plt.yticks(fontsize=12)

# Add titles and labels with increased font sizes
plt.title('Counts of Sequences by Country and Sequenced By', fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Counts', fontsize=14)

# Adjust legend with increased font size
plt.legend(title='Sequenced By', loc='upper right', fontsize=12, title_fontsize=14)

# Save the plot as an image file
plt.tight_layout()
plt.savefig('sequence_counts_by_country.png', dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_excel('/shares/grossniklaus.botinst.uzh/eharputluoglu/mmc3.xlsx', header=1)

# Group the data by Country and Sequenced by, then count occurrences
country_sequence_counts = data.groupby(['Country', 'Sequenced by']).size().reset_index(name='Counts')

# Pivot the data to wide format for easier analysis
pivot_data = country_sequence_counts.pivot(index='Country', columns='Sequenced by', values='Counts').fillna(0)

# Convert back to long format
long_data = pivot_data.reset_index().melt(id_vars='Country', var_name='Sequenced_by', value_name='Counts')

# Set up the matplotlib figure with increased size
plt.figure(figsize=(14, 10))

# Create a bar plot using seaborn with increased font sizes
sns.barplot(data=long_data, x='Country', y='Counts', hue='Sequenced_by', palette='viridis', ci=None)

# Rotate x-axis labels for better readability and increase font size
plt.xticks(rotation=45, ha='right', fontsize=12)

# Increase the font size for y-axis labels
plt.yticks(fontsize=12)

# Add titles and labels with increased font sizes
plt.title('Number of Samples per Country and Sequenced By', fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Samples', fontsize=14)

# Adjust legend with increased font size
plt.legend(title='Sequenced By', loc='upper right', fontsize=12, title_fontsize=14)

# Save the plot as an image file
plt.tight_layout()
plt.savefig('samples_per_country_and_sequenced_by.png', dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/shares/grossniklaus.botinst.uzh/eharputluoglu/accessions.csv')

# Group the data by Country and count the number of samples for each
country_sample_counts = data.groupby('Country').size().reset_index(name='Counts')

# Set up the matplotlib figure with increased size
plt.figure(figsize=(14, 10))

# Create a bar plot using seaborn with increased font sizes
sns.barplot(data=country_sample_counts, x='Country', y='Counts', palette='viridis')

# Rotate x-axis labels for better readability and increase font size
plt.xticks(rotation=45, ha='right', fontsize=12)

# Increase the font size for y-axis labels
plt.yticks(fontsize=12)

# Add titles and labels with increased font sizes
plt.title('Number of Samples per Country', fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Samples', fontsize=14)

# Save the plot as an image file
plt.tight_layout()
plt.savefig('samples_per_country.png', dpi=300)

# Show the plot
plt.show()
```
